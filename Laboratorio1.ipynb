```python
import numpy as pd
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
import keras
from keras.models import Sequential 
from keras.layers import Dense
from sklearn.metrics import accuracy_score
#KFolds
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
```

```python
dataset = pd.read_csv("D:/Jhonnatan/Documents/MAESTRIA DATA SCIENCE/3er Trimestre/Statical Learning III/TAREAS/titanic/train.csv")
```

```python
dataset.head()
```

```python
dataset.columns
```

```python
dataset = dataset[["Survived", "Pclass", "Sex", "Age", "SibSp","Parch", "Embarked"]]
```

```python
#verificacion de NAs
dataset.isnull().values.any()
```

```python
dataset.head()
```

```python
moda_edad = dataset.mode()
moda_edad[["Age"]]
```

```python
dataset = dataset.fillna(24)
```

```python
dataset = dataset[dataset.Age.notnull()]
dataset.head()
```

```python
ohe = pd.get_dummies(dataset['Sex'],drop_first = True)
dataset.drop('Sex', axis = 1, inplace = True)
dataset =  dataset.join(ohe)
dataset.head()
```

```python
ohe = pd.get_dummies(dataset['Embarked'])
dataset.drop('Embarked', axis = 1, inplace = True)
dataset =  dataset.join(ohe)
dataset.head()
```

```python
#verificacion de NAs
dataset.isnull().values.any()
```

```python
x = dataset.iloc[:, dataset.columns != 'Survived']
y = dataset.loc[:,'Survived']
```

```python
std_scaler = StandardScaler()
x = std_scaler.fit_transform(x)
x_train, x_test, y_train, y_test = train_test_split(x,y ,test_size =0.3 #,random_state = 2022
                                                    , shuffle = True)
```

```python
inputs = x_train.shape[1]

# base secuencial 
clasificador = Sequential()

#primera capa oculta

clasificador.add(Dense(input_dim = inputs, units = 6 , activation = 'relu', kernel_initializer = 'uniform'))
                 
#segunda capa oculta
                 
clasificador.add(Dense(units =6, activation ='relu', kernel_initializer ='uniform'))

#capa de salida
                  
clasificador.add(Dense(units=1, activation = 'sigmoid', kernel_initializer = 'uniform'))
```

```python
# parametros d eotpimizacion 

clasificador.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])


#parametros de entrenamiento

clasificador.fit(x_train, y_train, batch_size = 25, epochs = 150)
```

```python
#Guardar el modelo

path = 'D:/Jhonnatan/Documents/MAESTRIA DATA SCIENCE/3er Trimestre/Statical Learning III/TAREAS/titanic/'
clasificador.save(path+ '6_6_sig_out.h5')
```

```python

#cargamos a la read 

new_model = keras.models.load_model(path+ '6_6_sig_out.h5')
```

```python
y_pred = new_model.predict(x_test)
#y_pred
```

```python
y_pred = (y_pred >= 0.5)
y_pred
```

```python
accuracy_score(y_test, y_pred)
```
